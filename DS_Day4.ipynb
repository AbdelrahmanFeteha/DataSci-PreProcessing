```python
import pandas as pd
```

```python
units_sold = [50,80,60]
cost_per_unit = [10,12,9]
sales_price_per_unit = [15,18,12]

units_sold_df = pd.DataFrame(units_sold)
cost_per_unit_df = pd.DataFrame(cost_per_unit)
sales_price_per_unit_df = pd.DataFrame(sales_price_per_unit)

#make them all one dataframe
merged_df = pd.concat([units_sold_df, cost_per_unit_df, sales_price_per_unit_df], axis=1)

merged_df

total_profit = (sales_price_per_unit_df - cost_per_unit_df) * units_sold_df
total_profit = pd.DataFrame(total_profit)
merged_df = pd.concat([merged_df, total_profit], axis=1)
merged_df

#add column names
merged_df.columns = ['units_sold', 'cost_per_unit', 'sales_price_per_unit', 'total_profit']
merged_df
```

## Binning

```python
data = {'Age' : [18,25,47,51,62]}
df = pd.DataFrame(data)

bins = [0,18,35,50,100]
labels = ['Youth', 'Young Adult', 'Adult', 'Senior']
df['Age_Group'] = pd.cut(df['Age'], bins=bins, labels=labels)
df
```

## Task 1

###Load the data

```python
data = {
    'Age' : [22,25,47,51,62],
    'Salary' : [20000,25000,47000,51000,62000],
    'City' : ['New York', 'Los Angeles', 'Chicago', 'New York', 'Chicago']
}

df = pd.DataFrame(data)
df
```

### Feature Engineering

```python
##create a new fature age salary ratio (age / salary)
df['Age_Salary_Ratio'] = df['Age'] / df ['Salary']
df
```

### Min Max Scaling

```python
##min max scaling for salary
scaler = MinMaxScaler()
# Initialize the MinMaxScaler
scaler = MinMaxScaler()

#min max
df['Salary_Scaled'] = scaler.fit_transform(df[['Salary']])

#Faster method:
#columns_to_scale = ['Salary', 'Age',]
#f[columns_to_scale] = scaler.fit_transform(df[columns_to_scale])

# Print the scaled data
print("Min-Max Scaled data:\n", df)
```

### One Hot Encoding

```python
##Handle categorical variables

# Initialize OneHotEncoder
encoder = OneHotEncoder(sparse_output=False)

encoded_cities = encoder.fit_transform(df[['City']])

# Convert the encoded cities into a DataFrame
encoded_df = pd.DataFrame(encoded_cities, columns=encoder.get_feature_names_out(['City']))

# Concatenate the original DataFrame with the encoded DataFrame
df = pd.concat([df, encoded_df], axis=1)
df
```

### Dropping unecessary columns

```python
# Correctly dropping the unnecessary columns
df.drop(['Salary', 'City'], inplace=True, axis=1)

df
```
